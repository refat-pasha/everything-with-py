  # Defining Function Parameters in Python
  In Python, there are several ways to define function parameters:

  ## 1. Required Parameters
  These are parameters that must be provided when the function is called.

  ```python
  def greet(name):
      return f'Hello, {name}!'
  ```

  ## 2. Default Parameters
  These are parameters that take a default value if no value is provided.

  ```python
  def greet(name, greeting='Hello'):
      return f'{greeting}, {name}!'
  ```

  ## 3. Keyword Parameters
  Parameters can be passed by explicitly stating the parameter name.

  ```python
  greet(name='Alice', greeting='Hi')
  ```

  ## 4. Variable-Length Arguments
  You can use `*args` for variable-length positional arguments and `**kwargs` for keyword arguments.

  ```python
  def add_numbers(*args):
      return sum(args)

  def print_info(**kwargs):
      for key, value in kwargs.items():
          print(f'{key}: {value}')
  ```

# Expressions and Sequence Processing Functions in Python
Expressions and sequence processing functions play a crucial role in handling sequences such as lists and strings.

## Expressions
### 1. Indexing
Accessing elements in a sequence using their index.

```python
fruits = ['apple', 'banana', 'cherry']
print(fruits[1])  # Output: banana
```

### 2. Slicing
Extracting a portion of a sequence.

```python
numbers = [0, 1, 2, 3, 4, 5]
print(numbers[2:5])  # Output: [2, 3, 4]
```

### 3. Concatenation
Combining two sequences using the `+` operator.

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]
combined = list1 + list2
print(combined)  # Output: [1, 2, 3, 4, 5, 6]
```

## Sequence Processing Functions
### 1. `len()`
Returns the number of elements in a sequence.

```python
my_list = [10, 20, 30, 40]
print(len(my_list))  # Output: 4
```

### 2. `sum()`
Returns the sum of all elements in a numeric sequence.

```python
numbers = [1, 2, 3, 4, 5]
print(sum(numbers))  # Output: 15
```

### 3. `min()` and `max()`
Returns the minimum and maximum values in a sequence.

```python
scores = [88, 92, 95, 70, 85]
print(min(scores))  # Output: 70
print(max(scores))  # Output: 95
```

### 4. `sorted()`
Returns a new sorted list from the elements of any sequence.

```python
unsorted = [5, 2, 9, 1, 5, 6]
sorted_list = sorted(unsorted)
print(sorted_list)  # Output: [1, 2, 5, 5, 6, 9]
```

### 5. `join()`
Concatenates the elements of an iterable with a specified separator.

```python
words = ['Hello', 'world']
sentence = ' '.join(words)
print(sentence)  # Output: Hello world
```

````python
def calculate_total_cost():

    num_books = int(input("Enter the number of books: "))
    while num_books <= 0:
        print("Number of books cannot be zero or negative.")
        num_books = int(input("Please enter a valid number of books: "))


    total_cost = 0


    for i in range(1, num_books + 1):

        price = float(input(f"Price of book-{i}: "))

        while price < 0:
            print("Price cannot be negative.")
            price = float(input(f"Please enter a valid price for book-{i}: "))

        total_cost += price


    if 1 <= num_books <= 3:

        discount = 0
        discount_percentage = 0

    elif 4 <= num_books <= 10:

        discount_percentage = 10
        discount = total_cost * (discount_percentage/100)

    else:

        discount_percentage = 20
        discount = total_cost * (discount_percentage/100)


    payable_cost = total_cost - discount


    print(f"Total Cost: {total_cost:.2f}tk")

    if discount > 0:
        print(f"Total Payable Cost: {payable_cost:.2f}tk ({discount_percentage}% discount applied for purchasing {num_books} books)")
    else:
        print(f"Total Payable Cost: {payable_cost:.2f}tk (No discount applied)")


calculate_total_cost()
````
